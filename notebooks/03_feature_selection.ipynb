```python
import os
import sys
from pathlib import Path

import pandas as pd
import numpy as np
from IPython.display import clear_output

cwd = Path.cwd()  # Current directory
sys.path.append(str(cwd.parents[0]))
from feature_selection_function import perform_rfe,select_rfe_model
from definitions import LocationConfig
lc = LocationConfig()
```

```python
# Read train/test data for feature selection
df_raw_data = pd.read_csv(lc.feature_transformation_output_table)
```

```python
# Identiy non feature columns which needs to be removed 
columns_to_drop = []
target_variable = 'price_range' # Replace with your target column
```

```python
# Drop the columns which won't contribute to feature selection
df_raw_data = df_raw_data.drop(columns_to_drop, axis=1)
```

```python
# Differentiate Features and Target variable
X_features = df_raw_data.drop(target_variable, axis=1)  # Features
y_target = df_raw_data[target_variable]  # Target variable
```
